```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from numpy import linalg as LA
iris =  datasets.load_iris() 
x = iris.data
#print(x)
```

```python
x1=[]
x2=[]
x3=[]
x4=[]
for i in range(len(x)):
    x1.append(x[i][0])
    x2.append(x[i][1])
    x3.append(x[i][2])
    x4.append(x[i][3])
print(x1)
print(x2)
print(x3)
print(x4)
```

```python
x1m=sum(x1)/len(x1)
x2m=sum(x2)/len(x2)
x3m=sum(x3)/len(x3)
x4m=sum(x4)/len(x4)
print(x1m,x2m,x3m,x4m)
```

```python
xx1=[]
xx2=[]
xx3=[]
xx4=[]
for i in range(len(x1)):
    t=x1[i]-x1m
    xx1.append(t)
    t=x2[i]-x2m
    xx2.append(t)
    t=x3[i]-x3m
    xx3.append(t)
    t=x4[i]-x4m
    xx4.append(t)
print(xx1,xx2,xx3,xx4)
```

```python
A=np.array([xx1,xx2,xx3,xx4])
print(A)
```

```python
B=A.T
print(B)
```

```python
CM=np.dot(A,B)
CM=(1/(len(x)-1))*CM
print(CM)
```

```python
ev, E = LA.eigh(CM)
print(ev)
print(E)
```

```python
p=ev
mi1=p.argmax(axis=0)
p2=np.delete(p,mi1)
mi2=p.argmax(axis=0)
if mi2==mi1:
    mi2-=1
elif mi2>mi1:
    mi2+=1
print(mi1,mi2) 
print(E[mi1])
print(E[mi2])
```

```python
pc1=[]
pc2=[]
for i in range(len(x)):
    pc1.append(np.dot(B[i],E[mi1]))
    pc2.append(np.dot(B[i],E[mi2]))
print(pc1)
print()
print(pc2)
```

```python
total=sum(ev)
print(ev[mi1]*100/total ,'%')
print(ev[mi2]*100/total ,'%')
```

```python
final=np.array([pc1,pc2])
print(final)
```
