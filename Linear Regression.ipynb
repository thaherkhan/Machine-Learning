```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file=pd.read_csv('salary_data.csv')
```

```python
X = file.YearsExperience
Y = file.Salary
```

```python
m = 0
c = 0
l = 0.0001


for i in range(0,30000):
    yy = []
    for j in range(0,len(X)):
        yy.append(m*X[j]+c)
        
    dm = (-2/(len(X))) * sum(X*(Y-yy))
    dc = (-2/(len(X))) * sum(Y-yy)
    
    m = m-l*dm
    c = c-l*dc
    
print(m,c)
```

```python
print(yy)
```

```python
plt.scatter(X, Y) 
plt.scatter(X, yy, color='green')
plt.plot(X, yy, color='black')
plt.show()
```
